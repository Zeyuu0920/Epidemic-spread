Function used to calculate the node distance as edge_weight

```python
def calculate_distance(x1, y1, x2, y2):
    return np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
```

Function to calculate a full connected graph, which edge_weight is 1/(distance + k). k is a constant, used to avoid 0

```python
def abstract_graph(df, k=20):
    df_edge = df[["PN", "x.loc", "y.loc"]]
    num_nodes = len(df)

    G_full = nx.complete_graph(num_nodes)
    for i in range(num_nodes):
        for j in range(i+1, num_nodes):
            distance = calculate_distance(df_edge.iloc[i]["x.loc"], df_edge.iloc[i]["y.loc"], df_edge.iloc[j]["x.loc"], df_edge.iloc[j]["y.loc"])
            weight = 1 / (distance + k)
            G_full.add_edge(i, j, weight=weight)

    return G_full
```

Function to calculate the k_nearest_neighbors graph. What needs to be ensured is that the finalized graph structure is a fully connected graph

```python
def k_nearest_neighbors(G, neighbors):
    new_graph = nx.DiGraph()
    for node in enumerate(G.nodes()):
        new_graph.add_node(node[1])
        edges = list(G.edges(node, data=True))
        sorted_edges = sorted(edges, key=lambda x: x[2]["weight"],reverse=True)
        for num, i in enumerate(sorted_edges):

            if new_graph.has_edge(i[1], i[0]):
                continue

            if num == neighbors:
                break

            new_graph.add_edge(i[0],i[1],weight=i[2]["weight"])

    connected_components = nx.weakly_connected_components(new_graph)
    largest_component = max(connected_components, key=len)
    largest_subgraph = new_graph.subgraph(largest_component)
    return largest_subgraph
```

```python
def abstrct_edge_lst(G, min_neighbors=3):
    neighbors = min_neighbors - 1
    G_neighbor = nx.DiGraph()

    while len(G_neighbor.nodes()) < len(G.nodes()):
        neighbors += 1
        G_neighbor = k_nearest_neighbors(G,neighbors)

    connected_components = nx.weakly_connected_components(G_neighbor)
    num_connected_components = (sum(1 for _ in connected_components))

    assert num_connected_components == 1

    edge_list = list(G_neighbor.edges())
    edge_weights_dict = nx.get_edge_attributes(G_neighbor, 'weight')
    edge_weight_list = [edge_weights_dict[edge] for edge in edge_list]

    return edge_list, edge_weight_list
```

Functions for discrete variables with minmax normalization of results

```python
def discrete(df, column_name):
    if column_name == "C":
        serial_lst = (df[column_name] == "no complicatons").astype(int).tolist()
        return serial_lst


    if df[column_name].isna().any():
        serial_withoutNA = df[column_name].fillna(0)
        serial_num = pd.factorize(serial_withoutNA)[0].reshape(-1, 1)
    else:
        serial_num = pd.factorize(df[column_name])[0].reshape(-1, 1)
    scaler = MinMaxScaler()
    res = scaler.fit_transform(serial_num)
    serial_lst = res.flatten().tolist()
    return serial_lst
```

Functions for working with continuous variables and standardizing the results by standard deviation. Returns to scaler class when needed to recover predicted values

```python
def countinues(df, column_name, return_scaler=False):
    if df[column_name].isna().any():
        if column_name == "IFTO":
            data_list = df.groupby("NAME")[column_name].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x)).tolist()
    else:
        data_list = df[column_name].tolist()
    data_array = np.array(data_list).reshape(-1, 1)
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(data_array)
    scaled_lst = scaled_data.flatten().tolist()
    if return_scaler:
        return scaled_lst, scaler
    return scaled_lst
```

Collate all the features and return an ndarray

```python
def abstract_data(df):
    columns = ["FN", "HN", "AGE", "SEX", "CL", "IFTO", "SI", "C" ,"CA", "NI", "GE"]
    # FN
    # FN_lst = discrete(df, "FN")
    # HN
    HN_lst = discrete(df, "HN")
    # AGE
    age_lst = countinues(df, "AGE")
    # SEX
    SEX_lst = discrete(df, "SEX")
    # Class
    CL_lst = discrete(df, "CL")
    # IFTO
    # IFTO_lst = countinues(df, "IFTO")
    # SI
    SI_lst = countinues(df, "SI")
    # C
    C_lst = discrete(df, "C")
    # CA
    CA_lst = countinues(df, "CA")
    # NI
    NI_lst = countinues(df, "NI")
    # GE
    GE_lst = countinues(df, "GE")
    # TD
    # TM
    clean_data = np.array([HN_lst, age_lst, SEX_lst, CL_lst, SI_lst, C_lst, CA_lst, NI_lst, GE_lst]).T
    return clean_data
```

Collate the two targets and return an ndarray

```python
def abstract_label(df):
    label_column = ["tPRO", "tERU"]

    tPRO_lst, tPRO_scaler = countinues(df, "tPRO", return_scaler=True)
    tERU_lst, tERU_scaler = countinues(df, "tERU", return_scaler=True)

    label = np.array([tPRO_lst, tERU_lst]).T
    return label, [tPRO_scaler, tERU_scaler]
```

Organize the results into Data format in pyG using train_mask

```python
def dataload(data, label, edge_lst, edge_weight, node_num, fold=0, total_folds=5, whole=False):
    x = torch.tensor(data, dtype=torch.float32)
    y = torch.tensor(label, dtype=torch.float32)
    edge_index = torch.tensor(edge_lst, dtype=torch.long).t()
    edge_weight = torch.tensor(edge_weight, dtype=torch.float32)

    indices = torch.randperm(node_num)
    fold_size = node_num // total_folds
    test_idx = indices[fold * fold_size: (fold+1)*fold_size]

    remaining_idx = torch.cat([indices[:fold * fold_size], indices[(fold + 1) * fold_size:]])

    val_size = len(remaining_idx) // 4
    val_idx = remaining_idx[:val_size]
    train_idx = remaining_idx[val_size:]

    test_mask = torch.zeros(node_num, dtype=torch.bool)
    val_mask = torch.zeros(node_num, dtype=torch.bool)
    train_mask = torch.zeros(node_num, dtype=torch.bool)

    test_mask[test_idx] = True
    val_mask[val_idx] = True
    train_mask[train_idx] = True

    y_train = y[train_mask]
    y_val = y[val_mask]
    y_test = y[test_mask]

    if whole:
        return Data(x=x, edge_index=edge_index, y=y, edge_weight=edge_weight)

    return Data(x=x, edge_index=edge_index, y=y_train, edge_weight=edge_weight, target_mask=train_mask), \
           Data(x=x, edge_index=edge_index, y=y_val, edge_weight=edge_weight, target_mask=val_mask),\
           Data(x=x, edge_index=edge_index, y=y_test, edge_weight=edge_weight, target_mask=test_mask),
```

GCN model

```python
class GCN(torch.nn.Module):
    def __init__(self, num_features, num_targets, hidden_layers=[64, 64], dropout_rate=0.5):
        super().__init__()
        self.hidden_layers = hidden_layers
        self.dropout_rate = dropout_rate

        self.conv1 = GCNConv(num_features, self.hidden_layers[0])
        self.bn1 = BatchNorm(self.hidden_layers[0])

        self.convs = torch.nn.ModuleList()
        self.bns = torch.nn.ModuleList()
        for i in range(len(hidden_layers) - 1):
            self.convs.append(GCNConv(hidden_layers[i], hidden_layers[i + 1]))
            self.bns.append(BatchNorm(hidden_layers[i + 1]))

        self.conv_out = GCNConv(self.hidden_layers[-1], num_targets)

    def forward(self, data):
        x, edge_index, edge_weight = data.x.to(device), data.edge_index.to(device), data.edge_weight.to(device)
        x = F.relu(self.bn1(self.conv1(x, edge_index, edge_weight)))
        # x = F.dropout(x, p=self.dropout_rate, training=self.training)

        for conv, bn in zip(self.convs, self.bns):
            x = F.relu(bn(conv(x, edge_index, edge_weight)))
            x = F.dropout(x, p=self.dropout_rate, training=self.training)

        x = self.conv_out(x, edge_index, edge_weight)
        return x
```

Calculating RMSE

```python
def rmse_loss(input, target):
    return torch.sqrt(F.mse_loss(input, target))
```

```python
## edge_lst and edge_weight
df = read_file(raw_file)
G_full = abstract_graph(df)
edge_lst, edge_weight = abstrct_edge_lst(G_full)
```

```python
## Data
df = read_file(raw_file)
data = abstract_data(df)
```

```python
## Label
df = read_file(raw_file)
label, scaler_lst = abstract_label(df)
```

```python
whole_data = dataload(data, label, edge_lst, edge_weight, node_num, whole=True)
```

Training using 5-fold cross validation

```python
test_score_mse = []
test_score_rmse = []

test_score_mae = []
test_score_mape = []
test_score_medae = []

train_losses_lst = []
val_scores_lst = []

for fold in range(total_fold_num):
    train_data, val_data, test_data = dataload(data, label, edge_lst, edge_weight, node_num, fold, total_fold_num)

    model = GCN(num_features, num_targets).to(device)

    optimizer = torch.optim.Adam(model.parameters(), lr=lr)

    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=10, factor=0.5, min_lr=1e-10)

    loss_func = rmse_loss

    # train
    train_losses = []
    val_scores = []

    best_score = float("inf")
    stagnation = 0

    for epoch in range(Epochs):
        optimizer.zero_grad()

        out = model(train_data)
        loss = loss_func(out[train_data.target_mask], train_data.y.to(device))
        score = F.mse_loss(out[val_data.target_mask], val_data.y.to(device))

        loss.backward()
        optimizer.step()

        scheduler.step(score)

        train_losses.append(loss.item())
        val_scores.append(score.item())

        print(f"Epoch {epoch + 1}, Loss: {loss.item()}, Score: {score.item()}")

        if score.item() < best_val_score:
            best_val_score = score.item()
            torch.save(model.state_dict(), best_model_path)

        if score.item() < best_score:
            best_score = score.item()
            stagnation = 0
        else:
            stagnation += 1

        if stagnation == patience:
            print(f"No improvement for {patience} consecutive epochs, stopping training.")
            break

    train_losses_lst.append(train_losses)
    val_scores_lst.append(val_scores)

    model.eval()
    with torch.no_grad():
        pred = model(test_data)
        
        y_true = test_data.y.cpu().numpy()
        y_pred = pred[test_data.target_mask].cpu().numpy()
        
        mse = mean_squared_error(y_true, y_pred)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        
        mae = mean_absolute_error(y_true, y_pred)
        medae =  median_absolute_error(y_true, y_pred)

        test_score_mse.append(mse)
        test_score_rmse.append(rmse)
        
        test_score_mae.append(mae)
        test_score_medae.append(medae)
        
                
    print(f"MSE for fold_{fold+1}: {mse}")
    print(f"RMSE for fold_{fold+1}: {rmse}")

    print(f"MAE for fold_{fold+1}: {mae}")
    print(f"MedAE for fold_{fold+1}: {medae}")
```

Compute mse, rmse, mae, medae on training set

```python
test_score_mse_np = np.array(test_score_mse)
mse_mean_value = np.mean(test_score_mse_np)
print(f"MSE Mean value for 5 folds are {mse_mean_value}")


test_score_rmse_np = np.array(test_score_rmse)
rmse_mean_value = np.mean(test_score_rmse_np)
print(f"RMSE Mean value for 5 folds are {rmse_mean_value}")


test_score_mae_np = np.array(test_score_mae)
mae_mean_value = np.mean(test_score_mae_np)
print(f"MAE Mean value for 5 folds are {mae_mean_value}")

test_score_medae_np = np.array(test_score_medae)
medae_mean_value = np.mean(test_score_medae_np)
print(f"MedAE Mean value for 5 folds are {medae_mean_value}")
```
